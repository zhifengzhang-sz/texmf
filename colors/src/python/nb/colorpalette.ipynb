## Color palette generators
[Shades of Gray  Yes, Really](https://www.viget.com/articles/shades-of-gray-yes-really/)

### Quarters and thirds
This is to seperate grays into quarters and thirds as a way of covering a good range with memorable increments; but, it seems that the gaps werent equal. In other words, it ends up with shades that gather closer to the middle of the range than on the ends. When this happens, it misses some good contrasts on the lighter and darker sides of the range.

```python
end=spectra.html('#000000')
start=spectra.html('#ffffff')
scale=spectra.scale([start,end])
```

```python
swatches([scale(1),scale(0.5),scale(0)])
```

```python
swatches([scale(1),scale(.66),scale(.5),scale(.33),scale(0)])
```

```python
swatches([scale(1),scale(.75),scale(.66),scale(.5),scale(.33),scale(.25),scale(0)])
```

### Fixing the in-between
To fix this imbalance we can move the generator outward from a 50% middle with increments of 11%. This does a couple of nice things. Not only does it retain a perfect middle at 50% -- it goes further to create perfect middles at each gap in-between. It also gets two additional grays (17% and 83%) closer to each end of the spectrum for more options using lighter lights and darker darks. An added bonus is that it eliminates the pure blacks and whites which are nice to keep reserved for limited use, if any.

```python
swatches([scale(.94),scale(.5),scale(.06)])
```

```python
swatches([scale(.94),scale(.72),scale(.5),scale(.28),scale(.06)])
```

```python
swatches([scale(.94),scale(.83),scale(.72),scale(.61),scale(.5),scale(.39),scale(.28),scale(.17),scale(.06)])
```

### Warming and cooling
Of course I could stop there -- but that would be boring. I usually like to add a hint of color for further interest and distinction. The first step is to decide whether or not to warm or cool the set. In this example, Im cooling things down with a bright blue from the Viget color palette. Ill simply overlay the blue and create an opacity of 10% so that the grays are still visible and dominant through the transparent blue. Its hardly noticeable, but doing so nicely cools our grays with a subtle hint of blue as seen here:

```python
palette_gray=[scale(.94),scale(.83),scale(.72),scale(.61),scale(.5),scale(.39),scale(.28),scale(.17),scale(.06)]
alice=spectra.html(PaletteDB.vgetdotcom['alice'])
swatches([x.blend(alice,.05) for x in palette_gray])
```

```python
skyee=spectra.html(PaletteDB.vgetdotcom['skyee'])
swatches([x.blend(skyee,.1) for x in palette_gray])
```

```python
coral=spectra.html(PaletteDB.vgetdotcom['coral'])
swatches([x.blend(coral,.05) for x in palette_gray])
```

```python
rowan=spectra.html(PaletteDB.vgetdotcom['rowan'])
swatches([x.blend(rowan,.1) for x in palette_gray])
```

```python
kelly=spectra.html(PaletteDB.vgetdotcom['kelly'])
swatches([x.blend(kelly,.1) for x in palette_gray])
```

## The 60/30/10 rule
According to this rule, 60% of the color used should be the main color, 30% the secondary color, and 10% the accent color.

Examples:
- [Spendesk](https://www.spendesk.com/en-eu/) uses purple as their main color, white as their secondary color, and teal as their accent color.
- [Greenlist](https://greenlistapp.com/) uses green as their main color, white as their secondary color, and pink as their accent color.
- [Demuxed](https://2020.demuxed.com/) uses yellow as their main color, black as their secondary color, and white as their accent color.

An important note about the 60/30/10 rule is that it's really more of a guideline than a rule. For example, you may wish to incorporate more than three colors in your palette. The main takeaway is to focus on the primary color and use other colors to divide elements, create contrast, and highlight important features.

```python
from Colors.html import less_palette, css_palette
from Colors.io import write_palette_file
palette_names=['primary','secondary','tertiary','complement']
palette_qi_body_text={'background-color':palette_qi_body_text[0],'color':palette_qi_body_text[1]}
palette_qi_main={n:c for (n,c) in zip(palette_names,palette_qi_main)}
palette_qi_s1={n:c for (n,c) in zip(palette_names,palette_qi_s1)}
palette_qi_s2={n:c for (n,c) in zip(palette_names,palette_qi_s2)}
palette_qi_s3={n:c for (n,c) in zip(palette_names,palette_qi_s3)}
```

```python
write_palette_file('body_text',palette_qi_body_text,filename('body_text',num_color,dir_light),less_palette)
write_palette_file('main',palette_qi_main,filename('main',num_color,dir_light),less_palette)
write_palette_file('s1',palette_qi_s1,filename('s1',num_color,dir_light),less_palette)
write_palette_file('s2',palette_qi_s2,filename('s2',num_color,dir_light),less_palette)
write_palette_file('s3',palette_qi_s3,filename('s3',num_color,dir_light),less_palette)
```
